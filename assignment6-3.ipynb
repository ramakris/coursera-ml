```python
## Splitting the test_data by the labels
image_test_dog=image_test.filter_by(["dog"], "label")
image_test_cat=image_test.filter_by(["cat"], "label")
image_test_bird=image_test.filter_by(["bird"], "label")
image_test_auto=image_test.filter_by(["automobile"], "label")
```

```python
dog_cat_neighbors = cat_model.query(image_test_dog, k=1)
```

```python
print dog_cat_neighbors["distance"]
```

```python
dog_dog_neighbors = dog_model.query(image_test_dog, k=1)
```

```python
dog_bird_neighbors = bird_model.query(image_test_dog, k=1)
dog_auto_neighbors = auto_model.query(image_test_dog, k=1)
```

```python


dog_distances=graphlab.SFrame({"dog-automobile": dog_auto_neighbors["distance"],"dog-bird": dog_bird_neighbors["distance"]})
```

```python
dog_distances["dog-cat"] = dog_cat_neighbors["distance"]
dog_distances["dog-dog"]   =  dog_dog_neighbors["distance"]
```

```python
print dog_distances
```

```python
print dog_distances[0]
```

```python
def is_dog_correct(row):
    if row["dog-dog"] == min(row.values()):
        return 1
    else:
        return 0
```

```python
rows=dog_distances.apply(is_dog_correct)
```
